```python
import pandas as pd

# Read the songs.csv into a dataframe
songs_df = pd.read_csv("songs.csv", sep=";")

# We will create a list to hold the dictionary of audio features of each track, which can then be easily made into a dataframe and added to the songs dataframe
audio_feat_l = []

# This will be a placeholder for any rows we can't get data for
none_dict = {'danceability':None, 'energy':None, 'key':None, 'loudness':None, 'mode':None, 'speechiness':None, 'acousticness':None, 'instrumentalness':None, 'liveness':None, 'valence':None, 'tempo':None, 'type':None, 'id':None, 'uri':None, 'track_href':None, 'analysis_url':None, 'duration_ms':None, 'time_signature':None}

for url in songs_df['Track URL']:
    # Use the url to get the track id from the API
    try:
        track_id = sp.track(url, market=None)['id']
    except:
        # Sometimes it can't find the track id from the url, so in this case we create a row with None in each column
        print(url)
        audio_feat_l.append(none_dict)
    # Use id to find dictionary of audio features of song, and append it to the list
    else:
        audio_feat = sp.audio_features(tracks = [track_id])[0]
        audio_feat_l.append(audio_feat)

# Create a dataframe from the list of dictionaries and save it to a .csv
audio_feat_df = pd.DataFrame(audio_feat_l)
audio_feat_df.to_csv('audio_features.csv', index=False, sep= ";")
```
